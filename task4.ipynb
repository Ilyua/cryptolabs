```python
def brute_xor_one(string,limit=50):
    arr =[]
    
    hex_str = bytearray.fromhex(string)
    for key in st.ascii_letters:
        decoded = xor_one(key,hex_str)
        if (is_eng_text(decoded,lim=limit)):
            arr.append(key+'----'+decoded)
    return arr
```
